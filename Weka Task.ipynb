```python
import weka.core.jvm as jvm
jvm.start(max_heap_size="1024m",packages=True)
```

```python
import pandas as pd
dataset = pd.read_csv('SAPS2016_ED3409.csv',encoding='iso-8859-1',thousands=",")
final_data = dataset.iloc[:,3:].astype(int)
data2 = pd.read_csv('EDMedianGrossIncomeLDCommuter.csv',encoding='iso-8859-1')
data2 = (data2[data2['2016']!='..'])
data2['2016'].astype(int)
```

```python
final_data['MedianLDCommuter'] = data2['2016']
```

```python
data_dir = './'
```

```python
final_data = final_data[final_data['MedianLDCommuter'].notna()]
final_data.iloc[:,:].astype('category')
```

```python
columns = list(final_data)
for col in columns:
    final_data[col] = final_data[col].astype('category')
final_data.to_csv('./finaldataset2.csv',index=False)
```

```python
from weka.core.converters import Loader
loader = Loader(classname="weka.core.converters.CSVLoader")
data = loader.load_file(data_dir + "finaldataset2.csv")
data.class_is_last()

from weka.classifiers import Classifier
cls = Classifier(classname="weka.classifiers.trees.REPTree")

cls.build_classifier(data)

cls
```

```python
cls2 = Classifier(classname="weka.classifiers.rules.ZeroR")
cls2.build_classifier(data)
cls2
```

```python
data = loader.load_file(data_dir + "finaldataset2.csv")
data.delete_last_attribute()

from weka.clusterers import Clusterer
clusterer = Clusterer(classname="weka.clusterers.SimpleKMeans", options=["-N", "3"])
clusterer.build_clusterer(data)

print(clusterer)

# cluster the data
for inst in data:
    cl = clusterer.cluster_instance(inst)  # 0-based cluster index
    dist = clusterer.distribution_for_instance(inst)   # cluster membership distribution
    print("cluster=" + str(cl) + ", distribution=" + str(dist))
```

```python
from weka.classifiers import Classifier, Evaluation
from weka.core.classes import Random
from weka.classifiers import PredictionOutput, KernelClassifier, Kernel
from weka.filters import Filter
from weka.classifiers import Evaluation
from weka.core.classes import Random
data = loader.load_file(data_dir + "finaldataset2.csv")# previously loaded data
nominal = Filter(classname="weka.filters.unsupervised.attribute.NumericToNominal", options=["-R", "first-last"])
nominal.inputformat(data)
nominaldata = nominal.filter(data)
nominaldata.class_is_last()
classifier = Classifier(classname="weka.classifiers.trees.J48")
evaluation = Evaluation(nominaldata)          # initialize with priors
evaluation.crossvalidate_model(classifier, nominaldata, 10, Random(42))  # 10-fold CV
print(evaluation.summary())
print("pctCorrect: " + str(evaluation.percent_correct))
print("incorrect: " + str(evaluation.incorrect))
```
